```python
#plotting two arrays

arr1 = np.array([1,2,3,4,5,6,7,8])
arr2 = np.array([2,2,2,2,2,2,2,2])
c4 = np.array([2,3,4,6,7,8,6])
#increasing the s__sizeof__

plt.figure(figsize=(15,13)) #size
plt.plot(arr1,c='red' , linestyle="--" ,label="My Prediction") #linestyle
plt.plot(arr2, c='indigo', linewidth=5, label="Action") #colo.width
plt.plot(c4,c='blue' , linestyle="dashed", label="hhfjhfj" )
plt.title('RENT PER MONTH SALARY' ,fontsize=30,  fontweight="bold", ) #title

#after adding label call legend to plot 
plt.legend() #use loc="upper right or left  for location"

#to use month
plt.xticks(range(0,len(arr1)),('jan','feb','match','april','may','june','july','aug'))


#if to use numbevars

#plt.xticks([0,1,2,3,4,])

#adding title to y and x axis 
plt.xlabel('Month')
plt.ylabel("Rent $")
```

```python
#creting row data
category = ['ACCOUNT A','ACCOUNT B','ACCOUNT C',]
revenue = np.array([300,400,500])
```

```python
#plotting
plt.figure(figsize=(10,5))
```

```python
#bar
plt.bar(range(0,3),revenue)
plt.xticks(range(0,3),category)
```

```python
#line
plt.subplot(133)
plt.plot(range(0,3),revenue)
plt.xticks(range(0,3),category)
plt.figure(figsize=(70,70))

```

```python
#scatter
plt.scatter(range(0,3),revenue)
plt.xticks(range(0,3),category)
```
